```python
from transformers import AutoModelForCausalLM, AutoTokenizer, Trainer, TrainingArguments, BitsAndBytesConfig
from peft import LoraConfig
import bitsandbytes as bnb  # Importing bitsandbytes for quantization

# Load model with 4-bit quantization using the right approach
model = AutoModelForCausalLM.from_pretrained(
    "HuggingFaceH4/zephyr-7b-beta",
    quantization_config=BitsAndBytesConfig(load_in_4bit=True),  # Correct approach for 4-bit quantization
    device_map="auto",  # Distribute model on available devices (useful for large models)
)

tokenizer = AutoTokenizer.from_pretrained("HuggingFaceH4/zephyr-7b-beta")
```

```python
from datasets import load_dataset
import json

# Load dataset
dataset = load_dataset("allenai/SciRIFF", name="4096", split="train")

# Filter examples that have both input and output
filtered = dataset.filter(lambda x: x["input"] and x["output"])

# Select first 600 examples (or fewer if not enough)
filtered = filtered.select(range(min(600, len(filtered))))

# Write to JSONL
with open("sciriff_600_flat.jsonl", "w") as f:
    for example in filtered:
        json.dump({
            "input": example["input"],
            "output": example["output"]
        }, f)
        f.write("\n")
```

```python
from datasets import Dataset

# Assuming `filtered` is your dataset and has 'input' and 'output' columns
def preprocess_function(examples):
    # Tokenize inputs and outputs
    model_inputs = tokenizer(examples['input'], padding="max_length", truncation=True)
    labels = tokenizer(examples['output'], padding="max_length", truncation=True)

    # Return both input_ids and labels
    model_inputs['labels'] = labels['input_ids']
    return model_inputs

# Apply the preprocessing function to your dataset
train_dataset = filtered.map(preprocess_function, batched=True)
```

```python
def formatting_func(example):
    prompt = example["input"]
    response = example["output"]
    full_text = f"{prompt} {response}"  # simple concat; adjust if needed
    tokenized = tokenizer(full_text, truncation=True, padding="max_length", max_length=512)
    tokenized["labels"] = tokenized["input_ids"].copy()
    return tokenized
```

```python
tokenized_dataset = train_dataset.map(formatting_func, remove_columns=train_dataset.column_names)
```

```python
def format(example):
    prompt = example['input']
    response = example['output']
    full_text = f"{prompt} {response}"
    tokenized = tokenizer(full_text, truncation=True, padding="max_length", max_length=512)
    tokenized['labels'] = tokenized['input_ids'].copy()
    return tokenized
```

```python
tokenized_dataset = train_dataset.map(format, remove_columns=train_dataset.column_names)
```

```python
from peft import LoraConfig, get_peft_model, prepare_model_for_kbit_training
# Prepare for training
model = prepare_model_for_kbit_training(model)
```

```python
# Apply PEFT with LoRA
from peft import LoraConfig, get_peft_model
lora_config = LoraConfig(
    r=8,               # Low rank for LoRA (can adjust depending on model)
    lora_alpha=16,      # Scaling factor for LoRA (you can experiment with this)
    lora_dropout=0.05,   # Dropout rate for LoRA
    bias="none",        # No bias in LoRA layers
    task_type="CAUSAL_LM",  # Task type
    target_modules=["q_proj", "v_proj"]
)

# Apply the LoRA adapters to the quantized model
model = get_peft_model(model, lora_config)

# Tokenizer
tokenizer = AutoTokenizer.from_pretrained("HuggingFaceH4/zephyr-7b-alpha")

# Training arguments
training_args = TrainingArguments(
    output_dir="./results",
    per_device_train_batch_size=1,  # Adjust batch size as needed
    per_device_eval_batch_size=1,
    fp16=True,
    gradient_checkpointing=True,
    logging_dir="./logs",
    logging_steps=10,
    save_total_limit=2,
    save_strategy="steps",
    save_steps=500,
    eval_steps=500,
    disable_tqdm=False,
    num_train_epochs=1,  # Single epoch
    report_to="none" # <--- disables W&B
)

# Trainer using Hugging Face Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset,  # Your dataset
    tokenizer=tokenizer,
)

# Start training
trainer.train()
```

```python
# Save the adapter (LoRA) weights
model.save_pretrained("zephyr-finetuned-model")
tokenizer.save_pretrained("zephyr-finetuned-model")
```
